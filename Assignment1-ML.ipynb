```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.tree import plot_tree
from sklearn.datasets import load_iris, load_diabetes, load_breast_cancer
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import classification_report, confusion_matrix, explained_variance_score, mean_absolute_error
```

**Iris Data set**

```python
iris_data = load_iris()
X = iris_data.data
Y = iris_data.target
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30)

# Decision tree classifier(entropy optimization)

classifier = DecisionTreeClassifier(criterion="entropy")
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print("--------------------------Decision Tree Confusion matrix (entropy)--------------------------")
print(confusion_matrix(y_test, y_pred))

print("--------------------------Decision Tree Report (entropy)--------------------------")
print(classification_report(y_test, y_pred))

print("--------------------------Decision tree (entropy)--------------------------")
fig=plt.figure(figsize=(10, 10))
plot_tree(decision_tree=classifier, filled=True, feature_names=iris_data.feature_names)

# Decision tree classifier(gini optimization)

classifier = DecisionTreeClassifier(criterion="gini")
classifier.fit(x_train, y_train)

y_pred = classifier.predict(x_test)
print("--------------------------Decision Tree Confusion matrix (gini)--------------------------")
print(confusion_matrix(y_test, y_pred))

print("--------------------------Decision Tree Report (gini)--------------------------")
print(classification_report(y_test, y_pred))

print("--------------------------Decision tree (gini)--------------------------")
fig=plt.figure(figsize=(10, 10))
plot_tree(decision_tree=classifier, filled=True, feature_names=iris_data.feature_names)

# Naive Bayes classifier

# Multinomial NB
mnb = MultinomialNB().fit(x_train, y_train)
mnb.fit(x_train, y_train)

y_pred = mnb.predict(x_test)

print("--------------------------Multinomial Naive Bayes Confusion matrix--------------------------")
print(confusion_matrix(y_test, y_pred))

print("--------------------------Multinomial Naive Bayes Report--------------------------")
print(classification_report(y_test, y_pred))

# Gaussian NB
gnb = GaussianNB().fit(x_train, y_train)
gnb.fit(x_train, y_train)

y_pred = gnb.predict(x_test)

print("--------------------------Gaussian Naive Bayes Confusion matrix--------------------------")
print(confusion_matrix(y_test, y_pred))

print("--------------------------Gaussian Naive Bayes Report--------------------------")
print(classification_report(y_test, y_pred))

# Bernoulli NB
bnb = BernoulliNB().fit(x_train, y_train)
bnb.fit(x_train, y_train)

y_pred = bnb.predict(x_test)

print("--------------------------Bernoulli Naive Bayes Confusion matrix--------------------------")
print(confusion_matrix(y_test, y_pred))

print("--------------------------Bernoulli Naive Bayes Report--------------------------")
print(classification_report(y_test, y_pred))
```

**Diabetes data**

```python
diabetes_data = load_diabetes()

X = diabetes_data.data
Y = diabetes_data.target

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

linear_regression = LinearRegression()
linear_regression.fit(X_train, Y_train)
Y_pred = linear_regression.predict(X_test)

print("The model explains,", np.round(explained_variance_score(Y_test,Y_pred)*100,2),"% variance of the target")
print("The Mean Absolute Error of model is:", np.round(mean_absolute_error(Y_test,Y_pred ),2))
```
